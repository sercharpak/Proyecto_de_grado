### Cropping
The cropping is made by hand. This can definitively be done better.

```python
n_x, n_y = data.pixel_array.shape
```

```python
cut_prop = 0.09
cut_y_0 = 3.6
cut_x_0 = 3.4
cut_y_final = 4.35
cut_x_final = 3.75
n_colums_x = int(n_x*cut_prop)
n_colums_y = int(n_y*cut_prop)
```

```python
#Arrays of the column index to be deleted
array_index_columns_x_0 = np.arange(n_colums_x + cut_x_0*n_colums_x)
array_index_columns_y_0 = np.arange(n_colums_y + cut_y_0*n_colums_y)
array_index_columns_x_final = np.linspace(n_x-cut_x_final*n_colums_x,n_x-1, cut_x_final*n_colums_x)
array_index_columns_y_final = np.linspace(n_y-cut_y_final*n_colums_y,n_y-1, cut_y_final*n_colums_y)
```

```python
img_delete_x = np.delete(img,array_index_columns_x_final,0 )
img_delete_x = np.delete(img_delete_x,array_index_columns_x_0,0 )
img_delete_y = np.delete(img_delete_x,array_index_columns_y_final,1 )
img_delete_y = np.delete(img_delete_y,array_index_columns_y_0,1 )
```

```python
fig = plt.figure(figsize = (10,10))
plt.gray()
plt.grid()
plt.imshow(img_delete_y)
plt.xlabel("y(pixels)")
plt.ylabel("x(pixels)")
plt.show()
plt.close(fig)
```

```python
img = img_delete_y
n_x_new, n_y_new = img.shape
```

```python
cut_x_img_umbr = 0.35 #Va a umbralizar el 0.25 de la imagen independientemente que el 0.75
n_x_cut = int(n_x_new*cut_x_img_umbr)
```

## Umbralization
We umbralize the two sections of the aorta artery separately using otsu after equalizing the histogram.

```python
img_ascendent = np.zeros((n_x_new, n_y_new))
img_descendent = np.zeros((n_x_new, n_y_new))
```

```python
for j in range (n_y_new):
    for i in range(n_x_new):
        if(i<n_x_cut):
            img_ascendent[i,j] = img[i,j]
        else:
            img_descendent[i,j] = img[i,j]
```

```python
fig = plt.figure(figsize = (16,8))
fig.add_subplot(1, 2, 1)  # this line outputs images side-by-side
plt.imshow(img_ascendent, cmap = 'gray', origin='lower')
plt.title('Ascendent Aorta', fontsize=18)
plt.xlabel("x (pixels) ", fontsize=20)
plt.ylabel("y (pixels) ", fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=20)
fig.add_subplot(1, 2, 2)  # this line outputs images side-by-side
plt.imshow(img_descendent,  cmap = 'gray', origin='lower')
plt.title('Descendent Aorta', fontsize=18)
plt.xlabel("x (pixels) ", fontsize=20)
plt.ylabel("y (pixels) ", fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=20)
plt.show()
plt.close(fig)
```

```python
img_ascendent_binary, thresh_square = umbralizar_otsu(img_ascendent)
img_descendent_binary, thresh_coral = umbralizar_otsu(img_descendent)
```

```python
fig = plt.figure(figsize = (16,8))
fig.add_subplot(1, 2, 1)  # this line outputs images side-by-side
plt.imshow(img_ascendent_binary, cmap = 'gray', origin='lower')
plt.title('Ascendent Aorta umbralized', fontsize=18)
plt.xlabel("x (pixels) ", fontsize=20)
plt.ylabel("y (pixels) ", fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=20)
fig.add_subplot(1, 2, 2)  # this line outputs images side-by-side
plt.imshow(img_descendent_binary,  cmap = 'gray', origin='lower')
plt.title('Descendent Aorta umbralized', fontsize=18)
plt.xlabel("x (pixels) ", fontsize=20)
plt.ylabel("y (pixels) ", fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=20)
plt.show()
plt.close(fig)
```

```python
# Equalization
img_ascendent_eq = exposure.equalize_hist(img_ascendent)
# Equalization
img_descendent_eq = exposure.equalize_hist(img_descendent)
```

```python
fig = plt.figure(figsize = (16,8))
fig.add_subplot(1, 2, 1)  # this line outputs images side-by-side
plt.imshow(img_ascendent_eq, cmap = 'gray', origin='lower')
plt.title('Ascendent Aorta equalized', fontsize=18)
plt.xlabel("x (pixels) ", fontsize=20)
plt.ylabel("y (pixels) ", fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=20)
fig.add_subplot(1, 2, 2)  # this line outputs images side-by-side
plt.imshow(img_descendent_eq,  cmap = 'gray', origin='lower')
plt.title('Descendent Aorta equalized', fontsize=18)
plt.xlabel("x (pixels) ", fontsize=20)
plt.ylabel("y (pixels) ", fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=20)
plt.show()
plt.close(fig)
```

```python
img_ascendent_binary, thresh_square = umbralizar_otsu(img_ascendent_eq)
img_descendent_binary, thresh_coral = umbralizar_otsu(img_descendent_eq)
```

```python
fig = plt.figure(figsize = (16,8))
fig.add_subplot(1, 2, 1)  # this line outputs images side-by-side
plt.imshow(img_ascendent_binary, cmap = 'gray', origin='lower')
plt.title('Ascendent Aorta umbralized', fontsize=18)
plt.xlabel("x (pixels) ", fontsize=20)
plt.ylabel("y (pixels) ", fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=20)
fig.add_subplot(1, 2, 2)  # this line outputs images side-by-side
plt.imshow(img_descendent_binary,  cmap = 'gray', origin='lower')
plt.title('Descendent Aorta umbralized', fontsize=18)
plt.xlabel("x (pixels) ", fontsize=20)
plt.ylabel("y (pixels) ", fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=20)
plt.show()
plt.close(fig)
```

```python
img_thresholded = np.zeros((n_x_new, n_y_new))
for j in range (n_y_new):
    for i in range(n_x_new):
        if(i<n_x_cut):
            img_thresholded[i,j] = ~img_ascendent_binary[i,j]
        else:
            img_thresholded[i,j] = ~img_descendent_binary[i,j]
```

```python
fig = plt.figure(figsize = (10,10))
plt.gray()
plt.imshow(img_thresholded)
plt.xlabel("y(pixels)")
plt.ylabel("x(pixels)")
plt.show()
plt.close(fig)
```

```python
img_thresholded_float =  util.img_as_float(img_thresholded)
```

## Erosion
We erode several times the images to separate the aorta from the hearth.
At the end we will have to diluate the results the same number of times.

```python
n_dilation_erosion = 6
d = img_thresholded_float
```

```python
for i in range (n_dilation_erosion):
       d = morphology.erosion(d)
```

```python
fig = plt.figure(figsize = (10,10))
plt.gray()
plt.imshow(d)
plt.xlabel("y(pixels)")
plt.ylabel("x(pixels)")
plt.show()
plt.close(fig)
```

## Connected components
We get the conected components. We know that our aorta corresponds to the second and third larger groups.

```python
etiquetas, num = label(d, connectivity=2, return_num=True)
```

```python
print (num)
```

```python
fig = plt.figure(figsize = (10,10))
plt.gray()
plt.imshow(etiquetas)
plt.xlabel("y(pixels)")
plt.ylabel("x(pixels)")
plt.show()
plt.close(fig)
```

```python
final_ascendent = np.zeros(etiquetas.shape)
final_descendent = np.zeros(etiquetas.shape)
```

### Region Properties
<p>You can get a LOT of properties</p>
<p>http://scikit-image.org/docs/dev/api/skimage.measure.html#skimage.measure.regionprops</p>
<p>http://scikit-image.org/docs/dev/auto_examples/plot_label.html</p>

```python
for region in regionprops(etiquetas):
    print (region.area, region.label)
```

```python
for region in regionprops(etiquetas):
    if (region.label==2):
        for pos in region.coords:
            i=pos[0]
            j=pos[1]
            final_ascendent[i,j]=1.0
    if (region.label==7):
        for pos in region.coords:
            i=pos[0]
            j=pos[1]
            final_descendent[i,j]=1.0
```

```python
fig = plt.figure(figsize = (16,8))
fig.add_subplot(1, 2, 1)  # this line outputs images side-by-side
plt.imshow(final_ascendent, cmap = 'gray', origin='lower')
plt.title('Ascendent Aorta umbralized', fontsize=18)
plt.xlabel("x (pixels) ", fontsize=20)
plt.ylabel("y (pixels) ", fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=20)
fig.add_subplot(1, 2, 2)  # this line outputs images side-by-side
plt.imshow(final_descendent,  cmap = 'gray', origin='lower')
plt.title('Descendent Aorta umbralized', fontsize=18)
plt.xlabel("x (pixels) ", fontsize=20)
plt.ylabel("y (pixels) ", fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=20)
plt.show()
plt.close(fig)
```

## Dilution
We diluate to get the original areas

```python
for i in range (n_dilation_erosion):
    final_ascendent = morphology.dilation(final_ascendent)
    final_descendent = morphology.dilation(final_descendent)
```

```python
fig = plt.figure(figsize = (16,8))
fig.add_subplot(1, 2, 1)  # this line outputs images side-by-side
plt.imshow(final_ascendent, cmap = 'gray', origin='lower')
plt.title('Ascendent Aorta umbralized', fontsize=18)
plt.xlabel("x (pixels) ", fontsize=20)
plt.ylabel("y (pixels) ", fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=20)
fig.add_subplot(1, 2, 2)  # this line outputs images side-by-side
plt.imshow(final_descendent,  cmap = 'gray', origin='lower')
plt.title('Descendent Aorta umbralized', fontsize=18)
plt.xlabel("x (pixels) ", fontsize=20)
plt.ylabel("y (pixels) ", fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=20)
plt.show()
plt.close(fig)
```

```python
etiquetas_ascendent, num_ascendent = label(final_ascendent, connectivity=2, return_num=True)
etiquetas_descendent, num_descendent = label(final_descendent, connectivity=2, return_num=True)
```

## Radius estimation

```python
area_ascendent = 0.0
radius_ascendent = 0.0
for region_ascendent in regionprops(etiquetas_ascendent):
    print (region_ascendent.area, region_ascendent.label)
    area_ascendent = region_ascendent.area
    radius_ascendent = np.sqrt(area_ascendent/np.pi)
    print (radius_ascendent)
```

```python
area_descendent = 0.0
radius_descendent = 0.0
for region_descendent in regionprops(etiquetas_descendent):
    print (region_descendent.area, region_descendent.label)
    area_descendent = region_descendent.area
    radius_descendent = np.sqrt(area_descendent/np.pi)
    print (radius_descendent)
```

```python
#Taken from https://pyscience.wordpress.com/2014/09/08/dicom-in-python-importing-medical-image-data-into-numpy-with-pydicom-and-vtk/

# Load dimensions based on the number of rows, columns, and slices (along the Z axis)
ConstPixelDims = (int(data.Rows), int(data.Columns), len(fileset))

# Load spacing values (in mm)
ConstPixelSpacing = (float(data.PixelSpacing[0]), float(data.PixelSpacing[1]), float(data.SliceThickness))
```

```python
print (ConstPixelSpacing)
```

```python
radius_spacing = np.sqrt(ConstPixelSpacing[0]**2.0 + ConstPixelSpacing[1]**2.0)
```

```python
radius_ascendent_mm = radius_spacing * radius_ascendent
radius_descendent_mm = radius_spacing * radius_descendent
print (radius_ascendent_mm, radius_descendent_mm)
```

```python
print ("Our parameters for the medialness will then be, in mm:")
print ('$r_{max}$: ', radius_ascendent_mm*5.0/4.0)
print ('$r_{min}$: ', radius_descendent_mm/4.0)
```
